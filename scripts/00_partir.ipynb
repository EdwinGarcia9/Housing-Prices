```python
# Definir la ruta de la carpeta donde se guardarÃ¡n los archivos
carpeta_guardado = '../data/'

# Crear la carpeta si no existe
os.makedirs(carpeta_guardado, exist_ok=True)

# Definir el nÃºmero de partes en las que deseas dividir el df_usa
num_partes = 4

# Calcular el tamaÃ±o de cada parte
tamano_parte = len(df_usa) // num_partes

# Iterar sobre cada parte del df_usa y guardarla en un archivo separado
for i in range(num_partes):
    inicio = i * tamano_parte
    fin = (i + 1) * tamano_parte if i < num_partes - 1 else None
    parte = df_usa.iloc[inicio:fin]
    nombre_archivo = f'real_state_usa_{i + 1}.csv'
    ruta_archivo = os.path.join(carpeta_guardado, nombre_archivo)
    parte.to_csv(ruta_archivo, sep=';', index=False)

    # Puedes imprimir informaciÃ³n sobre la parte guardada
    print(f'Guardada parte {i + 1}: Filas de {inicio} a {fin} en {ruta_archivo}')
```
